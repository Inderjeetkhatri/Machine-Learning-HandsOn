```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x_input=np.array([1,3,5,8,9])
y_output=np.array([1,9,25,64,81])
z_output=np.array([81,64,25,9,1])
```

```python
plt.plot(x_input,y_output)
plt.show()
```

```python
plt.plot(x_input,y_output)
plt.xlabel("input")
plt.ylabel("output")
plt.title("x vs y")
plt.show()
```

```python
plt.plot(x_input,y_output,'r')
plt.xlabel("input")
plt.ylabel("output")
plt.title("x vs y")
plt.show()
```

```python
plt.plot(x_input,y_output,'k')
plt.xlabel("input")
plt.ylabel("output")
plt.title("x vs y")
plt.show()
```

```python
plt.plot(x_input,y_output,'o')
plt.xlabel("input")
plt.ylabel("output")
plt.title("x vs y")
plt.show()
```

```python
plt.plot(x_input,y_output,'bo')
plt.xlabel("input")
plt.ylabel("output")
plt.title("x vs y")
plt.show()
```

```python
#k=black
plt.plot(x_input,y_output,'ro')
plt.plot(x_input,y_output,'g')
plt.xlabel("input")
plt.ylabel("output")
plt.title("x vs y")
plt.show()
```

```python
plt.plot(x_input,y_output,'ro')
plt.plot(x_input,y_output,'g')
plt.plot(x_input,z_output,'bo')
plt.xlabel("input")
plt.ylabel("output")
plt.title("x vs y")
plt.show()
```

```python
plt.plot(x_input,y_output,'ro')
plt.plot(x_input,y_output,'g')
plt.plot(x_input,z_output,'*b')
plt.xlabel("input")
plt.ylabel("output")
plt.title("x vs y")
plt.legend(["y_output","z_output"])
plt.show()
```
